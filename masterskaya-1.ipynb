```python
# приведём столбцы к змеиному регистру
donor = donor.rename(columns={'ID':'id',
                              'Пол':'sex',
                              'Город':'city',
                              'Дата последней email активности':'e-mail_act'})

donor
```

```python
donat = donat.rename(columns={'Donor ID':'id',
                              'Дата платежа':'pay_date',
                              'Сумма платежа':'pay_sum',
                              'Рекуррентный':'recurrent',
                              'Назначение платежа':'purpose_of_payment',
                              'Канал оплаты':'payment_channel'})

donat
```

Кол-во уникальных значений одинаковое, поэтому соединять таблицы будем по этому столбцу.

```python
df = donor.merge(donat, how='left', on='id')
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
columns = ['id', 
           'sex', 
           'city', 
           'recurrent',
           'purpose_of_payment', 
           'payment_channel',
           'e-mail_act']

for i in columns:
    df[i] = df[i].fillna('')
 
df.isna().sum()
```

```python
df.head(10)
```

```python
df['pay_year'] = pd.DatetimeIndex(df['pay_date']).year
df['pay_month'] = pd.DatetimeIndex(df['pay_date']).month
df.info()
```

```python
df.head()
```

```python
df['pay_year'].unique()
```

```python
# удалить пробелы посмле слов в реккурентности
df['recurrent'] = df['recurrent'].replace(r'\s+','',regex=True)
```

```python
df.head()
```

```python
df['recurrent'].unique()
```

```python
df_rec = df.query("recurrent == 'Да'")
df_rec.head()
```

```python
df_fk = df.drop(columns=['sex', 'city',	'e-mail_act',	'purpose_of_payment',	'payment_channel',	'pay_month'], axis=1)
```

```python
df_fk.head()
```

```python
df_fk_d = df_fk.pivot_table(index='id', values='pay_date', aggfunc=['max', 'min']).reset_index()
```

```python
df_fk_d.columns=['id', 'max_date', 'min_date']
```

```python
df_fk_d.head()
```

```python
df_fk_d['duration'] = df_fk_d['max_date'] - df_fk_d['min_date']
df_fk_d.info()
```

```python
df_fk_d['duration'] = (df_fk_d['duration']).dt.days
```

```python
df_fk_d.head()
```

```python
df_fk_d['duration'].unique()
```

```python
df_fk_d.info()
```

```python
df_fk_d['duration'].describe()
```

```python
df_fk_p = df_fk.pivot_table(index='id', values='pay_sum', aggfunc=['sum', 'count']).reset_index()
```

```python
df_fk_p.columns=['id', 'sum_pay', 'count_pay']
```

```python
df_fk_p.info()
```

```python
df_fk_p.head()
```

```python
df_fk_all = df_fk_d.merge(df_fk_p, on=['id'], how='outer')
```

```python
df_fk_all.head()
```

```python
df_fk_alll = df_fk_all.drop(columns=['min_date', 'max_date', 'id'])
```
